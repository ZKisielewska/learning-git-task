````python
# import necessary tools

# offline vizualization
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
init_notebook_mode(connected=True)

# import cufflinks library
import cufflinks as cf
cf.go_offline()

# import plotly objects
import plotly.offline as pyo
import plotly.graph_objs as go
from plotly import subplots

# import pandas library
import pandas as pd
````

````python
def configure_plotly_browser_state():
  import IPython
  display(IPython.core.display.HTML('''
        <script src="/static/components/requirejs/require.js"></script>
        <script>
          requirejs.config({
            paths: {
              base: '/static/base',
              plotly: 'https://cdn.plot.ly/plotly-1.5.1.min.js?noext',
            },
          });
        </script>
        '''))

configure_plotly_browser_state()
````

````python
kghm = pd.read_csv('https://uploads.kodilla.com/bootcamp/pro-data-visualization/files/kgh_d.csv')
kghm.head()
````

````python
copper = pd.read_csv('https://uploads.kodilla.com/bootcamp/pro-data-visualization/files/ca_c_f_d.csv')
copper.head()
````

````python
# remove selected columns from kghm table
kghm.drop(['Otwarcie', 'Najwyzszy', 'Najnizszy', 'Wolumen'], axis=1, inplace=True)
````

````python
# remove selected columns from copper table
copper.drop(['Otwarcie', 'Najwyzszy', 'Najnizszy'], axis=1, inplace=True)
````

````python
# merge the tables and rename the columns
share_prices = pd.merge(kghm, copper, on='Data', suffixes=('_kghm', '_copper'))
share_prices.rename(columns={'Data': 'Date', 'Zamkniecie_kghm': 'KGHM', 'Zamkniecie_copper': 'Copper'}, inplace=True)
share_prices.head()
````

````python
# create subplots
configure_plotly_browser_state()
fig = subplots.make_subplots(rows=2, cols=1, subplot_titles=['KGHM', 'Copper'], shared_xaxes=True)

fig.add_trace(go.Scatter(x=share_prices['Date'],
                         y=share_prices['KGHM'],
                         name='KGHM share prices'),1,1)

fig.add_trace(go.Scatter(x=share_prices['Date'],
                         y=share_prices['Copper'],
                         name='Copper prices'),2, 1)
````
